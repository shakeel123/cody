## Generere fil med Faker
Kodens overordnede formål er at generere en fil ("dummy_data.txt") med 10.000 rækker falske data, hvor hver række består af et falsk navn, telefonnummer, e-mailadresse og adresse. Dataene er formateret og skrevet til filen på en struktureret måde.

```python
from faker import Faker

fake = Faker()
with open("dummy_data.txt", "w") as file:
    for _ in range(10000):  # Change the number of rows as needed
        # Replace any newline characters in the address with spaces
        address = fake.address().replace('\n', ' ')
        
        # Create the data line
        line = f"{fake.name()} | {fake.phone_number()} | {fake.email()} | {address}\n"
        
        # Write the line to the file
        file.write(line)
```

```python
def read_file(file_path):
    with open(file_path, "r") as file:
        for line in file:
            yield line.strip().split(" | ")

# Example of using the generator
file_path = "dummy_data.txt"
data_generator = read_file(file_path)

for data in data_generator:
    # Check if the line has enough components
    if len(data) == 4:
        name, phone_number, email, address = data
        print(f"Name: {name}, Phone: {phone_number}, Email: {email}, Address: {address}")
    else:
        print(f"Invalid line: {data}")
```

### Filtering og data behandling: 2
Dette program læser igen filen ("dummy_data.txt") og behandler hver linje. Det tjekker først, om linjen har fire komponenter (navn, telefonnummer, e-mail, adresse). Hvis det er tilfældet, udpakker det disse komponenter og anvender filtreringslogik: det udskriver kun de linjer, hvor adressen indeholder ordet "Street". Derefter udfører det behandlingslogik: det fjerner ikke-numeriske tegn fra telefonnumre og udskriver det endelige resultat. Hvis en linje ikke har fire komponenter, udskriver den en fejlmeddelelse.

```python
with open("dummy_data.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    data = line.strip().split(" | ")

    # Check if the line has enough components
    if len(data) == 4:
        name, phone_number, email, address = data

        # Filtering logic: Only print entries where the address contains the word "Street"
        if "Street" in address:
            # Processing logic: Remove non-numeric characters from phone numbers
            processed_phone_number = ''.join(char for char in phone_number if char.isdigit())
            
            print(f"Name: {name}, Phone: {processed_phone_number}, Email: {email}, Address: {address}")
    else:
        print(f"Invalid line: {line}")
```

### Data Framework ved brug af Pandas:
las od bare konvertere denne txt. file til Pandas data frame ved brug af Pandas library. og dette funktionalities forfra. Pandas en kraftfuld datavidenskabsbibliotek i Python, der gør det lettere at arbejde med og analysere data i tabelformat.

```python
import pandas as pd

def process_data_to_dataframe(file_path):
    data_list = []

    with open(file_path, 'r') as file:
        for line in file:
            data = line.strip().split(" | ")

            # Check if the line has enough components
            if len(data) == 4:
                name, phone_number, email, address = data

                # Perform your custom logic here
                # For example, let's filter out entries where the name does not start with 'A'
                if not name.startswith('A'):
                    # Append the data to the list
                    data_list.append({
                        'Name': name,
                        'Phone': phone_number,
                        'Email': email,
                        'Address': address
                    })

    # Convert the list of dictionaries to a pandas DataFrame
    df = pd.DataFrame(data_list)
    
    return df

# Replace 'dummy_data.txt' with the actual path to your generated data file
file_path = 'dummy_data.txt'
df = process_data_to_dataframe(file_path)

# Set the index starting from 1
df.index = df.index + 1
# Display the DataFrame
df
```
